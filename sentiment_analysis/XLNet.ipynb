```python
from transformers import XLNetTokenizer, XLNetForSequenceClassification, pipeline
import torch
import csv
from transformers import AutoModel

# Function to read a column from a CSV file
def read_column(csv_file, column_name):
    data = []
    with open(csv_file, 'r', newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
            # Check if the column exists in the row
            if column_name and "Misc" in row:
                # Handle line breaks within the column
                column_body = row[column_name].replace('\n', ' ')
                try:
                  column_sentiment = int(row["Misc"])
                except:
                  print("Could not parse integer in Misc Column")
                data.append([column_body,column_sentiment])
    return data

def sentiment_from_label(label):
    return 1 if label == 'LABEL_1' else 0

access_token = "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"
csv_file = "/content/drive/MyDrive/sentiment_analysis_labeled_amazon.csv"
column_name = "Body"
column_data = read_column(csv_file, column_name)

sentiment_analysis = pipeline("text-classification", model="Ibrahim-Alam/finetuning-xlnet-base-cased-on-Cornell_sentiment", token=access_token)

tot_positive = 0
tot_negative = 0
guess_positive = 0
guess_negative = 0

true_positive = 0
true_negative = 0
false_positive = 0
false_negative = 0

for item in column_data:
  try:
    sentiment = sentiment_analysis(item[0])
    # print(sentiment)
    if sentiment_from_label(sentiment[0].get('label')) == 1:
      guess_positive += 1
    else:
      guess_negative += 1

    actual_sentiment = item[1]
    if actual_sentiment == 1:
      tot_positive += 1
    else:
      tot_negative += 1

    if actual_sentiment == 1 and sentiment_from_label(sentiment[0].get('label')) == 1:
      true_positive += 1
    elif actual_sentiment == 0 and sentiment_from_label(sentiment[0].get('label')) == 0:
      true_negative += 1
    elif actual_sentiment == 0 and sentiment_from_label(sentiment[0].get('label')) == 1:
      false_positive += 1
    elif actual_sentiment == 1 and sentiment_from_label(sentiment[0].get('label')) == 0:
      false_negative += 1

    print("GUESS: ", sentiment_from_label(sentiment[0].get('label')), " ACTUAL: ", actual_sentiment )
  except:
    print("Could not read sentiment from column data")

print()
print("TOTAL POSITIVE: ", tot_positive, "  GUESSED POSITIVE: ", guess_positive)
print("TOTAL NEGATIVE: ", tot_negative, "  GUESSED NEGATIVE: ", guess_negative)
print()
print("TRUE POSITIVE: ", true_positive, "  FALSE POSITIVE: ", false_positive)
print("FALSE NEGATIVE: ", false_negative, "  TRUE NEGATIVE: ", true_negative)
```

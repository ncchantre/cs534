```python
from yahoofinancials import YahooFinancials

az = YahooFinancials('AMZN').get_historical_price_data(start_date='2022-01-01', end_date='2022-12-31', time_interval='daily')
```

```python
print(az['AMZN']['prices'])
```

```python
import pandas as pd

df = pd.DataFrame(az['AMZN']['prices'])
print(df.head())

df.to_csv('amazon_stock_prices.csv', index=False)
```
